```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator, FuncFormatter
from scipy.interpolate import spline
font = { 'size' : 14}

matplotlib.rc('font', **font)
%matplotlib notebook
```

## Plot loss vs. epoch

```python
epochs = np.array(range(1, 5 + 1), dtype=np.float32)
losses_text8 = np.array([0.945, 0.712, 0.665, 0.631, 0.601])
losses_europarl = np.array([0.938, 0.848,0.775, 0.704, 0.655])

ax = plt.figure().gca()
ax.xaxis.set_major_locator(FixedLocator(epochs))
plt.plot(epochs, losses_text8, label = "text8", marker="s", markerfacecolor="none", markersize=10)
plt.plot(epochs, losses_europarl, label = "europarl", marker="s", markerfacecolor="none", markersize=10)
plt.xlabel("Epoch")
plt.ylabel("Average loss")
plt.legend()
#plt.show()
plt.savefig("plot1.pdf", bbox_inches="tight")
```

## Plot accuracy vs. epoch

```python
epochs = np.array(range(1, 5 + 1), dtype=np.float32)
text8 = np.array([0.112, 0.124, 0.124, 0.256, 0.296])
europarl = np.array([0.108, 0.124, 0.164, 0.164, 0.172])

ax = plt.figure().gca()
ax.xaxis.set_major_locator(FixedLocator(epochs))
plt.plot(epochs, text8, label = "text8", marker="s", markerfacecolor="none", markersize=10)
plt.plot(epochs, europarl, label = "europarl", marker="s", markerfacecolor="none", markersize=10)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
#plt.show()
plt.savefig("plot2.pdf", bbox_inches="tight")
```

## Plot accuracy vs. number of negative samples

```python
epochs = [1, 2, 5]
text8 = np.array([0.228, 0.216, 0.296])
europarl = np.array([0.188, 0.184, 0.172])

ax = plt.figure().gca()
ax.xaxis.set_major_locator(FixedLocator(epochs))
plt.plot(epochs, text8, label = "text8", marker="s", markerfacecolor="none", markersize=10)
plt.plot(epochs, europarl, label = "europarl", marker="s", markerfacecolor="none", markersize=10)
plt.xlabel("Number of negative samples")
plt.ylabel("Accuracy")
plt.legend()
#plt.show()
plt.savefig("plot3.pdf", bbox_inches="tight")
```
